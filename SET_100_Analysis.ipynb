```python
import pandas as pd

from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser
```

```python
output_string = StringIO()
with open('SET50_100_H1_2021_revised.pdf', 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, output_string, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    
    for pageNumber, page in enumerate(PDFPage.get_pages(in_file)):
        if pageNumber == 1:
            interpreter.process_page(page)

print(output_string.getvalue())
```

```python
#check numeric value https://stackoverflow.com/questions/19859282/check-if-a-string-contains-a-number
import re

def hasNumbers(input_str):
    return bool(re.search(r'\d', input_str))

#PCL: Public Company Limited (Thailand)
def hasPCL(input_str):
    return bool('PCL' in input_str)
```

```python
#show the element without PCL word or the numeric value - should give list of sectors
sector_list = [x for x in index_list if not hasPCL(x) and not hasNumbers(x)]
```

```python
#remove  the last 3 elements from the list 
index_list = index_list[:len(index_list) - 3]
sector_list = sector_list[:len(sector_list) - 3]
sector_list
```

```python
#give the set of sector name
set(sector_list)
```

```python
#remove few elements from the list
removal_ele = ['PSL', 'Sector', 'Rank by alphabet']
sector_list = [x for x in sector_list if x not in removal_ele]
sector_list = list(set(sector_list))
sector_list
```

```python
#remove the reserve list of SET100 and the sector from the index list 
removal_ele = ['PSL', 'Sector', 'Rank by alphabet', '1 AAV', '4', '2 ERW', '3 MC', '5 SIRI', '6 TKN', 
               'ASIA AVIATION PCL', 'THE ERAWAN GROUP PCL', 'MC GROUP PCL', 'PRECIOUS SHIPPING PCL', 'SANSIRI PCL', 
               'TAOKAENOI FOOD & MARKETING PCL', 'Sector', 'Rank by alphabet', 'SET100 Inclusion: OR', 
               'SET100 Exclusion: TKN', 'Reserve list of SET100 ']
index_list = [x for x in index_list if x not in removal_ele and x not in sector_list]
index_list
```

```python
#now the text should follow the pattern such that Symbol name followed by the full company as the next element
symbol_list = []
company_list = []
for idx, val in enumerate(index_list):
    if hasNumbers(val) and not hasPCL(val):
        symbol_list.append(val)
        company_list.append(index_list[idx+1])
```

```python
def transform_symbol(input_str):
    input_str = input_str.replace('#', '')
    remove_digits = str.maketrans('', '', digits)
    input_str = input_str.translate(remove_digits).replace(" ", "")
    return input_str + '.BK' #.BK for SET 
```

```python
import requests
from bs4 import BeautifulSoup

#Extract the sector and industry from the yahoofinance
def get_company_sector_industry(symbol):
    URL = 'https://finance.yahoo.com/quote/'+ symbol+ '/profile?p=' + symbol
#     print(symbol, URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find_all('span', class_='Fw(600)')
    if len(results) > 0:
        sector = results[0].text.strip()
        industry = results[1].text.strip()
    else:
        sector = ''
        industry = ''
    print(symbol, sector, industry)
    return sector, industry
```

```python
df['sector'],  df['industry'] = zip(*df['symbol'].apply(get_company_sector_industry))
```
